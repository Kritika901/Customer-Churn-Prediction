# Customer Churn — Exploratory Data Analysis

This notebook explores the IBM Telco Customer Churn dataset: distributions, relationships between features and churn, and correlations. Run `python src/preprocessing.py` first (or place the raw CSV in `data/raw/`) before running this notebook.

```python
import sys
sys.path.append('..')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

from src.preprocessing import run_preprocessing

sns.set_style('whitegrid')
pd.set_option('display.max_columns', None)
```

## 1. Load and Clean Data

```python
df = run_preprocessing(save=True)
df.head()
```

```python
df.info()
df.describe()
```

## 2. Churn Distribution

```python
fig = px.pie(df, names='Churn', title='Churn Distribution', hole=0.4)
fig.show()
```

## 3. Gender Distribution vs Churn

```python
fig = px.histogram(df, x='gender', color='Churn', barmode='group', title='Gender vs Churn')
fig.show()
```

## 4. Contract Type Analysis

```python
fig = px.histogram(df, x='Contract', color='Churn', barmode='group', title='Contract Type vs Churn')
fig.show()
```

## 5. Monthly Charges Analysis

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='Churn', y='MonthlyCharges')
plt.title('Monthly Charges by Churn')
plt.show()
```

## 6. Tenure Analysis

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='tenure', hue='Churn', kde=True, bins=30)
plt.title('Tenure Distribution by Churn')
plt.show()
```

## 7. Correlation Heatmap

```python
numeric_df = df.select_dtypes(include=[np.number])
plt.figure(figsize=(8,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

## 8. Pairplot of Numeric Features

```python
sns.pairplot(df, vars=['tenure', 'MonthlyCharges', 'TotalCharges'], hue='Churn', diag_kind='kde')
plt.show()
```

## 9. Interactive Plotly Scatter: Tenure vs Monthly Charges

```python
fig = px.scatter(df, x='tenure', y='MonthlyCharges', color='Churn', opacity=0.6,
                  title='Tenure vs Monthly Charges')
fig.show()
```

## 10. Key Takeaways

- Month-to-month contracts show substantially higher churn than long-term contracts.
- Customers with higher monthly charges (often fiber optic + streaming add-ons) churn more.
- Low-tenure customers are the highest-risk segment.
- Lack of Online Security / Tech Support correlates with higher churn.
